```python
import random
import csv
import numpy as np
import pandas as pd
def generate_matrix(length, min_value, max_value, seed):
    random.seed(seed)
    matrix = []
    for i in range(length):
        row = []
        for j in range(length):
            if i == j:
                row.append(0)
            else:
                row.append(random.randint(min_value, max_value))
        matrix.append(row)
    return matrix

def export_matrix_to_csv(matrix, filename):
    # Create a copy of the matrix with diagonal values set to 9999
    matrix_with_9999 = []
    for i in range(len(matrix)):
        row = []
        for j in range(len(matrix[i])):
            if i == j:
                row.append(9999)
            else:
                row.append(matrix[i][j])
        matrix_with_9999.append(row)

    # Write the copied matrix to a CSV file
    with open(filename, mode='w', newline='') as file:
        writer = csv.writer(file)
        writer.writerows(matrix_with_9999)
    print(f"Matrix has been exported to {filename}")

# User input
length = 11
min_value = 100
max_value = 999
j= 0
sd=42
matrices=[]
while j < 4:
    filename = input("Enter the filename (with .csv extension): ")

    # Generate the matrix
    matrix = generate_matrix(length, min_value, max_value, seed=sd)
    print("Random Square Matrix with Zero Diagonal:")
    for row in matrix:
        print(row)
    matrices.append(matrix)

    # Export the matrix to a CSV file with diagonal values set to 9999
    export_matrix_to_csv(matrix, filename)
    sd+=1
    j+=1
```

```python
DistMatrix=np.array(
        # fmt: off
        [[0,808,654,275,134,357,545,236,123,484,645],
      [808,0,156,724,752,860,960,870,849,1036,1207],
      [654,156,0,571,596,710,854,731,701,889,1063],
      [275,724,571,0,141,151,796,503,396,331,511],
      [134,752,596,141,0,242,663,364,255,396,570],
      [357,860,710,151,242,0,900,593,478,181,360],
      [545,960,854,796,663,900,0,313,429,1023,1171],
      [236,870,731,503,364,593,313,0,116,710,859],
      [123,849,701,396,255,478,429,116,0,595,746],
      [484,1036,889,331,396,181,1023,710,595,0,179],
      [645,1207,1063,511,570,360,1171,859,746,179,0],
        # fmt: on
    ]
)
DistMatrix=(DistMatrix/100)
demands=[0,40,56,64,11,92,38,53,64,5,75]
for i in range(len(demands)):
  demands[i]=int(np.round(demands[i]/100*524,0))
```

```python
%%time
#importar librerías
import psutil
# Your code here
print('RAM usage:', psutil.Process().memory_info().rss / 1024 ** 2, 'MB')

def calcular_distancia_total(ruta, matriz_distancias):
    """
    Calcular la distancia total de una ruta dada utilizando la matriz de distancias.
    """
    distancia_total = 0
    num_puntos = len(ruta)

    # Sumar las distancias entre cada par de nodos consecutivos en la ruta
    for i in range(num_puntos - 1):
        nodo_actual = ruta[i]
        siguiente_nodo = ruta[i + 1]
        distancia_total += matriz_distancias[siguiente_nodo, nodo_actual]

    return distancia_total

def vecino_mas_cercano(matriz_distancias, demandas, capacidad):
    """
    Aplicar la heurística del Vecino Más Cercano para encontrar rutas iniciales para el VRP.
    """
    tiempo_acumulado = 0
    num_puntos = matriz_distancias.shape[0]
    visitados = np.zeros(num_puntos, dtype=bool)  # Array para marcar los nodos visitados
    rutas = []
    for i in range(len(visitados)):
      if demandas[i]==0:
        visitados[i]=True


    # Mientras no se hayan visitado todos los nodos
    while np.sum(visitados) < num_puntos:
        nodo_actual = 0  # Empezar en el nodo 0 (depósito)
        capacidad_actual = 0
        ruta = [nodo_actual]
        visitados[nodo_actual] = True  # Marcar el depósito como visitado

        # Mientras la capacidad actual más la demanda del nodo actual no exceda la capacidad del vehículo
        while capacidad_actual + demandas[nodo_actual] <= capacidad:
            actual = ruta[-1]
            mas_cercano = None
            min_dist = float('inf')

            # Encontrar el vecino no visitado más cercano que cumpla con la capacidad del vehículo
            for vecino in np.where(~visitados)[0]:
                if demandas[vecino] + capacidad_actual <= capacidad and matriz_distancias[actual, vecino] < min_dist:
                    mas_cercano = vecino
                    min_dist = matriz_distancias[actual, vecino]

            # Si no se encuentra un vecino adecuado, regresar al depósito
            if mas_cercano is None:
                tiempo_acumulado += matriz_distancias[actual][0]
                break

            # Sumar la distancia al vecino más cercano y verificar si el tiempo acumulado excede 8 horas
            #Se le agrega el costo por planta de carga y descarga. Carga= 15 min/524 plantas=0.03. Descarga = 15 min/524 plantas =0.03. Carga+ Descarga = 0.06 min/planta
            tiempo_acumulado += (matriz_distancias[actual][mas_cercano]+0.06*demandas[vecino])


            # Añadir el vecino más cercano a la ruta y marcarlo como visitado
            ruta.append(mas_cercano)
            visitados[mas_cercano] = True
            capacidad_actual += demandas[mas_cercano]


        # Regresar al depósito y añadir la ruta a la lista de rutas
        ruta.append(0)
        rutas.append(ruta)

    return rutas

def formatear_output(rutas):
    """
    Formatear las rutas finales según se requiera.
    En este ejemplo, devuelve una lista de rutas.
    """
    return rutas

def solver_vrp(matriz_distancias, demandas, capacidad):
    """
    Resolver el VRP utilizando la heurística del Vecino Más Cercano.
    """
    rutas = vecino_mas_cercano(matriz_distancias, demandas, capacidad)
    rutas_formateadas = formatear_output(rutas)
    return rutas_formateadas



capacidad = 524

solucion = solver_vrp(DistMatrix, demands, capacidad)

costos = []
volumen=[]
# Imprimir la solución y calcular el costo total de cada ruta
for ruta in solucion:

    if ruta != [0,0]:
      print("RUTA " + str(solucion.index(ruta) + 1) + ": " + str(ruta))
      print()
      costos.append(0)
      volumen.append(0)
      for i in range(len(ruta) - 1):
          print("de: c" + str(ruta[i]))
          print("a: c" + str(ruta[i + 1]))

          costos[solucion.index(ruta)] += (DistMatrix[ruta[i]][ruta[i + 1]]+0.06*demands[ruta[i+1]])
          print("COSTO DEL PASO: " + str(DistMatrix[ruta[i]][ruta[i + 1]]+0.06*demands[ruta[i+1]]))
          volumen[solucion.index(ruta)] += demands[ruta[i+1]]
          print("VOLUMEN DEL PASO: " + str(demands[ruta[i+1]]))

      print("TIEMPO TOTAL DE LA RUTA: " + str(costos[solucion.index(ruta)]))
      print("VOLUMEN TOTAL DE LA RUTA: " + str(volumen[solucion.index(ruta)]))
      print()

    else:
      costos.append(0)
      volumen.append(0)
      print()
      costos[solucion.index(ruta)] =0

      volumen[solucion.index(ruta)] =0
      print("Termino de Jornada Laboral")
      print()

print("TIEMPO TOTAL PARA TODAS LAS RUTAS: " + str(sum(costos)))
costos_ex=costos
print("VOLUMEN TOTAL PARA TODAS LAS RUTAS: " + str(sum(volumen)))
```
